```python
%matplotlib inline
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load train and test data
train = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
train = train.drop(['PassengerId', 'Ticket'], axis = 1) # drop PassengerId and Ticket
y = train['Survived']
# train = train.drop('Survived', axis = 1) 
print train
```

```python
train[train.dtypes[train.dtypes == 'object'].index].describe() # to explore 'object' data types
```

```python
# Data Exploration 
train['Sex'] = pd.Categorical(train['Sex']) # covert Sex and Embarked into categorical variables
train['Embarked'] = pd.Categorical(train['Embarked'])
new_pclass = pd.Categorical(train['Pclass'], ordered = True)

# create a new feature representing the social class of the passengers on board
train['Pclass'] = new_pclass.rename_categories(['Class1','Class2','Class3'])
train['Pclass'].describe()
```

```python
train['Cabin'] = pd.Categorical(train['Cabin'].astype(str).str[0])
train['Cabin'].describe()
```

```python
train.hist(column = 'Age', figsize = (9,6), bins = 20) # the distribution of age
```

```python
# # median_value = train['Age'].median()
train['Age'] = train['Age'].fillna(train['Age'].median())
# train['Age'].describe()

# create age bands/bins
train['Age_Band'] = pd.cut(train['Age'], 5)
train[['Age_Band', 'Survived']].groupby('Age_Band', as_index = False).mean().sort_values(by = 'Age_Band', ascending = True)
train.loc[train['Age'] <= 16, 'Age'] = 0
train.loc[(train['Age'] > 16) & (train['Age'] <= 32), 'Age'] = 1
train.loc[(train['Age'] > 32) & (train['Age'] <= 48), 'Age'] = 2
train.loc[(train['Age'] > 48) & (train['Age'] <= 64), 'Age'] = 3
train.loc[train['Age'] > 64, 'Age'] = 4
train.drop(['Age_Band'], axis = 1)

# create Fare band/bins
train.loc[train['Fare'] <= 7.91, 'Fare'] = 0
train.loc[(train['Fare'] > 7.91) & (train['Fare'] <= 14.454), 'Fare'] = 1
train.loc[(train['Fare'] > 14.454) & (train['Fare'] <= 31), 'Fare'] = 2
train.loc[train['Fare'] > 31, 'Fare'] = 3
train['Fare'] = train['Fare'].astype(int)

# age_guess = np.zeros((2,3))
train.head(10)
```

```python
# visualizing the relationship between Survival and different features (predictor variables)
g = sns.FacetGrid(train, col = 'Survived')
g.map(plt.hist, 'Age', bins = 20)

n_g = sns.FacetGrid(train, col = 'Survived', row = 'Pclass', size = 2.2, aspect = 1.6)
n_g.map(plt.hist, 'Age', alpha = .5,bins = 20)
n_g.add_legend()

grid = sns.FacetGrid(train, row = 'Embarked', size = 2.2, aspect = 1.6)
grid.map(sns.pointplot, 'Pclass', 'Survived', 'Sex', palette  = 'deep')
grid.add_legend()

n_g2 = sns.FacetGrid(train, col = 'Survived', row = 'Embarked', size = 2.2, aspect = 1.6)
n_g2.map(sns.barplot, 'Sex', 'Fare', alpha = .5, ci = None)
n_g2.add_legend()
```

```python
# create a new feature called Family by adding SibSp and Parch
train['Family'] = train['SibSp'] + train['Parch']
train[train['Family'] == max(train['Family'])] # find the people with the max number of families on board
```

```python
# Encode all the categorical variables
categorical_data = pd.DataFrame([train['Sex'], train['Embarked'], train['Cabin'], train['Pclass']]).T
new_categorical_data = pd.get_dummies(categorical_data, columns = ['Sex', 'Embarked', 'Pclass', 'Cabin'])
new_categorical_data.head(5)
```

```python
new_train_data = pd.concat([train, new_categorical_data], axis = 1)

new_train_data = new_train_data.drop(['Name','Survived','Pclass', 'Sex', 'Embarked', 
                                      'Cabin', 'Cabin_T', 'Age_Band'], axis = 1)
```

```python
# using Logistic Regresssion 
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
model = LogisticRegression()
model.fit(new_train_data, y)
```

```python
prediction = model.predict(new_train_data)
print metrics.classification_report(y, prediction)
print 'Logistic Regression Accuracy:', round(metrics.accuracy_score(y, prediction), 2)

# the line below obtains the column names
dt_coeffs = pd.DataFrame(new_train_data.columns.delete(0))
dt_coeffs.columns = ['Feature']
dt_coeffs['Correlation'] = pd.Series(model.coef_[0])

dt_coeffs.sort_values(by = 'Correlation', ascending = True)
# plot ROC curve
fpr, tpr, _ = metrics.roc_curve(y, prediction)
roc_auc = metrics.auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color = 'darkorange', lw =2)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
# using Decision Trees
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(new_train_data, y)
dt_prediction = dt_model.predict(new_train_data)
print metrics.classification_report(y, dt_prediction)
print 'Decision Tree Accuracy', round(metrics.accuracy_score(y, dt_prediction), 2)
```

```python
# using Random Forest 
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(new_train_data, y)
rf_prediction = rf_model.predict(new_train_data)
print metrics.classification_report(y, rf_prediction)
metrics.accuracy_score(y, rf_prediction)
```

```python
# using SVM
from sklearn.svm import SVC
svm_model = SVC()
svm_model.fit(new_train_data, y)
svm_prediction = svm_model.predict(new_train_data)
print metrics.classification_report(y, svm_prediction)
print 'SVM Accuracy: ', round(metrics.accuracy_score(y, svm_prediction), 2)
```

```python
# training with cross validation 
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
cv_dt_model = DecisionTreeClassifier()
scores = cross_val_score(cv_dt_model, new_train_data, y, cv = 5)
scores
```
